```python
import tkinter as tk
import tkinter.messagebox as MessageBox

root = tk.Tk()
root.title("TIC-TAC-TOE")

root.geometry('400x400')

digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]

mark = ''

count = 0

panels = ['panel', 1, 2, 3, 4, 5, 6, 7, 8, 9]

def checker(digit):
    global count, mark, digits, panels, buttons, labels
    
    if digit in digits:
        digits.remove(digit)
        
        if count % 2 == 0:
            mark = 'X'
        else:
            mark = 'O'

        panels[digit] = mark
        buttons[digit-1].config(text=mark)

        count += 1

        if win(mark):
            MessageBox.showinfo("Game Over", f"Player {mark} wins!")
            root.destroy()
            return

        if count > 8:
            MessageBox.showinfo("Game Over", "Match Tied!")
            root.destroy()

def win(current_mark):
    win_conditions = [
        [1, 2, 3], [4, 5, 6], [7, 8, 9],
        [1, 4, 7], [2, 5, 8], [3, 6, 9],
        [1, 5, 9], [3, 5, 7] 
    ]
    
    for condition in win_conditions:
        if panels[condition[0]] == panels[condition[1]] == panels[condition[2]] == current_mark:
            return True
    return False

buttons = []
labels = []

current_player_label = tk.Label(root, text="Current Player: X", font=('GIL SANS MT', 18))
current_player_label.grid(row=0, column=0, columnspan=3)

for i in range(1, 10):
    button = tk.Button(root, text=str(panels[i]), fg = 'black', bg = 'yellow', width=14, height=6, font =('Times New Roman',12),
                       command=lambda i=i: checker(i))
    button.grid(row=(i-1)//3 + 1, column=(i-1)%3)
    buttons.append(button)

def update_current_player_label():
    if count % 2 == 0:
        current_player_label.config(text="Current Player: X")
    else:
        current_player_label.config(text="Current Player: O")

root.mainloop()
```
